**4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range**

> For each datasets we used the same techiquies in order to exctract the last date data. 

```python
times_max_year.reset_index(inplace=True)
times_max_year.head()
```

```python
shanghai_max_year = shanghai.sort_values(by=["year"],ascending=False)
shanghai_max_year = shanghai_max_year.drop_duplicates(["university_name"],keep="first")
shanghai_max_year.head()
```

```python
cwur_max_year = cwur.sort_values(by=["year"],ascending=False)
cwur_max_year = cwur_max_year.drop_duplicates(["university_name"],keep="first")
cwur_max_year.head()
```

> Sistemo le colonne per prepararle alla join

```python
times_rank = times_max_year.filter(['university_name', 'world_rank'])
times_rank.rename(columns={"world_rank":"world_rank_times"},inplace=True)
```

```python
shan_rank = shanghai_max_year.filter(['university_name', 'world_rank'])
shan_rank.rename(columns={"world_rank":"world_rank_shan"},inplace=True)
```

```python
cwur_rank = cwur_max_year.filter(['university_name', 'world_rank'])
cwur_rank.rename(columns={"world_rank":"world_rank_cwur"},inplace=True)
```

```python
# join three datasets on the university_name attribute
dfs= [times_rank,shan_rank,cwur_rank]
dfs = [df.set_index('university_name') for df in dfs]
rank_eval = dfs[0].join(dfs[1:],how="inner") 
```

```python
rank_eval.reset_index(inplace=True)
rank_eval
```

```python
# cleaning algorithm for "-,=" values in the ranking columns
for j in range(1,3):
    for i in range(0,len(rank_eval)):
        if rank_eval.iloc[i,j].find("-") != -1: 
            x = rank_eval.iloc[i,j].split("-")
            rank_eval.iloc[i,j] = np.mean([float(x[0]),float(x[1])])
        elif rank_eval.iloc[i,j].find("=") != -1:
                x = rank_eval.iloc[i,j].split("=")
                x = x[1]
                rank_eval.iloc[i,j] = float(x)
        else: rank_eval.iloc[i,j] = float(rank_eval.iloc[i,j])
```

```python
result = [] 
for index,row in rank_eval.iterrows():
    result.append(max([row["world_rank_times"],row["world_rank_shan"],row["world_rank_cwur"]])-min([row["world_rank_times"],row["world_rank_shan"],row["world_rank_cwur"]]))
    
rank_eval["diff_max_min"]=result
```

**5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.**

```python
times_max_year_5 = times_max_year
```

```python
times_max_year_5.dropna(inplace=True)
# missing value removed 
times_max_year_5.drop(axis=0,index=788,inplace=True)
```

```python
perc_female = []
perc_male= []
for i in range(0,len(times_max_year_5)):
    if times_max_year_5.iloc[i,12].find(":") != -1: 
        x = times_max_year_5.iloc[i,12].split(":")
        perc_female.append(float(x[0])/100)
        perc_male.append(float(x[1])/100)
    else: print(i)
```

```python
times_max_year_5["perc_female"],times_max_year_5["perc_male"] = perc_female,perc_male
```

```python
times_max_year_5
```

```python
for i in range(0,len(times_max_year_5)):
    times_max_year_5.iloc[i,9] = float(times_max_year_5.iloc[i,9].replace(',',''))
```

```python
# apply (?) forse c'√® un modo per farle contemporaneamente
times_max_year_5["num_female"] = times_max_year_5.apply(lambda x: (x["perc_female"]*x["num_students"]),axis=1)
times_max_year_5["num_male"] = times_max_year_5.apply(lambda x: (x["perc_male"]*x["num_students"]),axis=1)
```

```python
#newtable
times_max_year_5.drop(columns=["female_male_ratio"],inplace=True)
times_max_year_5
```

```python
# group by country and sum 
result_5 = times_max_year_5.groupby(by="country").sum()[["num_male","num_female"]]
result_5
```
