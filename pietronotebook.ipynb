**1. For each university, extract from the times dataset the most recent and the least recent data, obtaining two separate dataframes**

```python
times_min_year = times.sort_values(by=["year"],ascending=True)
times_min_year = times_min_year.drop_duplicates(["university_name"], keep='first')
```

**2. For each university, compute the improvement in income between the least recent and the most recent data points**

```python
# clean the income colomn in order to evaluate differences between them.
missing_values = ["n/a", "na", "-", "--"]
for i in range(0,len(times_max_year["income"])):
    if times_min_year.iloc[i,7] in missing_values:
        times_min_year.iloc[i,7] = np.nan
    if times_max_year.iloc[i,7] in missing_values:
        times_max_year.iloc[i,7] = np.nan
```

```python
times_min_year.set_index("university_name",inplace=True)
times_min_year
```

```python
time_join = times_max_year.join(other = times_min_year,how="inner",lsuffix="_max",rsuffix="_min",on="university_name")
```

```python
time_join.reset_index(inplace=True)
time_join
```

```python
rows =[] 
for index,row in time_join.iterrows():
    diff = float(row["income_max"])-float(row["income_min"])
    rows.append([row["university_name"],row["income_max"],row["income_min"],diff])
```

```python
time_income_diff = pd.DataFrame(rows,columns=["university_name","income_max_year","income_min_year","income_diff"])
```

**4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range**

```python
times_max_year.reset_index(inplace=True)
times_max_year.head()
```

```python
cwur_max_year = cwur.sort_values(by=["year"],ascending=False)
cwur_max_year
```

```python
cwur_max_year = cwur_max_year.drop_duplicates(["university_name"],keep="first")
cwur_max_year
```

```python
# join three datasets on the university_name attribute
dfs= [times_max_year,shanghai_max_year,cwur_max_year]
df = dfs[0]
for df_ in dfs[1:]:
    df = df.merge(df_, on='university_name')
```

```python
df.rename({"world_rank_x":"world_rank_times","world_rank_y":"world_rank_shanghai","world_rank":"world_rank_cwur"},axis=1,inplace=True)
rank_eval = df[["university_name","world_rank_times","world_rank_shanghai","world_rank_cwur"]]
```

```python
rank_eval
```

```python
# cleaning algorithm for "-,=" values in the ranking columns
for j in range(1,3):
    for i in range(0,len(rank_eval)):
        if rank_eval.iloc[i,j].find("-") != -1: 
            x = rank_eval.iloc[i,j].split("-")
            rank_eval.iloc[i,j] = np.mean([float(x[0]),float(x[1])])
        elif rank_eval.iloc[i,j].find("=") != -1:
                x = rank_eval.iloc[i,j].split("=")
                x = x[1]
                rank_eval.iloc[i,j] = float(x)
        else: rank_eval.iloc[i,j] = float(rank_eval.iloc[i,j])
```

```python
result = [] 
for index,row in rank_eval.iterrows():
    result.append(max([row["world_rank_times"],row["world_rank_shanghai"],row["world_rank_cwur"]])-min([row["world_rank_times"],row["world_rank_shanghai"],row["world_rank_cwur"]]))
    
rank_eval["diff_max_min"]=result
```
