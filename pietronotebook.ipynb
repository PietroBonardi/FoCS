### Queries

**1. For each university, extract from the times dataset the most recent and the least recent data, obtaining two separate dataframes**

```python
times.head()
```

```python
max_year_uni = times.groupby(by="university_name").max()    
```

```python
max_year_uni
```

```python
min_year_uni = times.groupby(by="university_name").min()
min_year_uni
```

**2. For each university, compute the improvement in income between the least recent and the most recent data points**

```python
for i in range(0,len( max_year_uni["income"])):
    if max_year_uni.iloc[i,6] in "-":
        max_year_uni.iloc[i,6] = np.nan
        
for i in range(0,len(min_year_uni["income"])):
    if min_year_uni.iloc[i,6] in "-":
        min_year_uni.iloc[i,6] = np.nan
```

```python
join_tab = max_year_uni.join(other=min_year_uni, how="inner",lsuffix="_maxyear",rsuffix="_minyear",on="university_name")
```

```python
join_tab.reset_index(inplace=True)
join_tab
```

```python
rows =[] 
for i in range(0,len(join_tab)):
    diff = float(join_tab.iloc[i,7])-float(join_tab.iloc[i,20])
    rows.append([join_tab.iloc[i,0],diff])
```

```python
result = pd.DataFrame(rows,columns=["University","income_diff"])
```
