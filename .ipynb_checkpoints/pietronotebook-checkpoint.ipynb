```python
# University with the most recent data
max_year = max(times["year"]) 

max_year_university = times[times["year"]== max_year]
max_year_university.groupby(by="university_name").sum()
```

```python
# University with the most recent data
min_year = min(times["year"])

min_year_university = times[times["year"]== min_year]
min_year_university.groupby(by="university_name").sum()
```
